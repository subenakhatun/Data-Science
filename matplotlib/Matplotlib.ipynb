### BAr chart
i. Bivariate analysis <br>
ii. categorical vs numerical <br>
iii. Aggregate(sum, average,min, max,div) analysis of groups

```python
# simple bar chart
import matplotlib.pyplot as plt
children = [10,5,30,50,15]
color =["red","green","yellow","black","white"]
plt.bar(color,children, color="red")
plt.title("Childern vs color")
plt.xlabel("color")
plt.ylabel("children")
```

```python
# horzontal: barh
children = [10,5,30,50,15]
color =["red","green","yellow","black","white"]
plt.barh(color,children, color="red")
plt.title("Childern vs color")
plt.xlabel("color")
plt.ylabel("children")
```

```python
df = pd.read_csv("datasets-session-23/batsman_season_record.csv")
df
```

```python
# Multiple bar chart. side by side as like building

plt.bar(np.arange(df.shape[0])-0.2,df["2015"], width=0.2)
plt.bar(np.arange(df.shape[0]),df["2016"],width=0.2)
plt.bar(np.arange(df.shape[0])+0.2,df["2017"],width=0.2)
plt.xticks(np.arange(df.shape[0]),df["batsman"])
```

```python
plt.bar(np.arange(df.shape[0])-0.2,df["2015"], width=0.2, color="green")
plt.bar(np.arange(df.shape[0]),df["2016"],width=0.2,color="red")
plt.bar(np.arange(df.shape[0])+0.2,df["2017"],width=0.2,color="pink")
plt.xticks(np.arange(df.shape[0]),df["batsman"], rotation="vertical")
plt.show()
```

```python
# Stack bar chart
plt.bar(df["batsman"],df["2015"], width=0.2, color="green", label="2015")
plt.bar(df["batsman"],df["2016"],bottom=df["2015"],width=0.2,color="red",label="2016")
plt.bar(df["batsman"],df["2017"],bottom=df["2015"] + df["2016"],width=0.2,color="yellow",label="2017")
plt.xticks(df["batsman"], rotation="vertical")
plt.legend()
plt.show()
```
