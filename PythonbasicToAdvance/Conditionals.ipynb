```python
# 11.Check if a number is odd or even

num = int(input("Enter a number: "))
if (num % 2) == 0:
    print(f"{num} is a even number.")
if (num % 2) == 1:
    print(f"{num} is a Odd number.")
else:
    print("Please give a number for check")
```

```python
# 12.Check if a number is positive, negative, or zero
num = int(input("Enter a number: "))
if num < 0:
    print(f"{num} is a Negative number.")
elif num > 0:
    print(f"{num} is a Positive number.")
else:
    print(f"{num} is a Zero.")    
```

```python
# 13.Find the largest of 3 numbers
num1 = int(input("Enter a number: "))
num2 = int(input("Enter a number: "))
num3 = int(input("Enter a number: "))
if num1>num2 and num1>num3:
    print(f"Number 1: {num1} is a Gretest number..")
elif num2>num1 and num2>num3:
    print(f"Number 2: {num2} is a Gretest number..")
else:
    print(f"Number 3: {num3} is a Gretest number..")
```

```python
# 14.	Check if a year is a leap year
leaf_year = int(input("Enter a year: "))
if leaf_year % 4 == 0:
    print(f"{leaf_year} is a Leaf year")
elif leaf_year%100 == 0:
    if leaf_year % 400 == 0:
        print(f"{leaf_year} is a Leaf year")
else:
    print(f"{leaf_year} is not a Leaf year")
```
