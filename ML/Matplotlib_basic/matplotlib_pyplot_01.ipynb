```python
# just import matplotlib and start its inline
# pyplot object
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# 2 numpy array same length .
x = [1,2,3,4,5,6,7]
y = [50,51,52,48,47,49,46]
# just insure your dataset or numpy arry is the same length
print("show the lenagth of x and y :", len(x),len(y))
```

```python
# show this data in pyplot grap . the result show line graph

plt.plot(x,y)
```

```python
# we can see its with x axis and y axis label name with lebel
# you change line style with linestyle propert use and linewidth
plt.xlabel("Day")
plt.ylabel("Temperature")
plt.plot(x,y, color="green", linewidth = 4, linestyle = "dotted")
```

```python
# use different style, color , and title 
plt.xlabel("Day")
plt.ylabel("Temperature")
plt.title("Temperature Graph")
plt.plot(x,y, "r+", linewidth = 4, markersize = 20, alpha=0.8)
```
