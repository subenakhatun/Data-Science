```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# four difference lists 
days = [1,2,3,4,5,6,7]
max_temp = [50,51,52,48,47,49,46]
min_temp = [43,42,40,44,33,35,37]
avg_temp = [45,48,48,46,40,42,41]
```

```python
plt.plot(days,max_temp)
```

```python
# want to represnt 3 different graph using this data
# days with max temperature
plt.xlabel("Days")
plt.ylabel("Weather")
plt.title("Weather Temperature 3 graph in one")
plt.plot(days,max_temp, label ="Max")
# days with min temperature
plt.plot(days,min_temp, label = "Min")
# days with average temperature
plt.plot(days,avg_temp, label = "Avg")
plt.legend()
# you can change legend loc postion using loc="upper right / upper left", shadow=True, fontsize ="large/small"
```

```python
# days with max temperature with grid 
plt.xlabel("Days")
plt.ylabel("Weather")
plt.title("Weather Temperature 3 graph in one")
plt.plot(days,max_temp, label ="Max")
# days with min temperature
plt.plot(days,min_temp, label = "Min")
# days with average temperature
plt.plot(days,avg_temp, label = "Avg")
plt.legend(loc="upper center", shadow=True, fontsize="large")
plt.grid()
# you can change legend loc postion using loc="upper right / upper left", shadow=True, fontsize ="large/small"
```
