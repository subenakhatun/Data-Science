```python
# import all important modules
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
%matplotlib inline
```

```python
# draw a pi chart
read_csv = pd.read_csv("expense.csv")
read_csv.head(5)
```

```python
ev = read_csv["Expense value"]
ev
```

```python
el = read_csv["expanse levels"]
el
```

```python
plt.pie(ev,labels=el)
plt.show()
```

```python
# draw a perfect circle
plt.pie(ev,labels=el, shadow=True)
plt.axis("equal")
plt.show()
```

```python
# show percentage for every pie
plt.axis("equal")
plt.pie(ev,labels=el, shadow=True, autopct="%1.1f%%", radius=1.5)
plt.show()
```

```python
# use explode
plt.axis("equal")
plt.pie(ev,labels=el, shadow=True, autopct="%1.1f%%", radius=1.5, explode=[0.3,0,0.2,0,0])
plt.show()
```
