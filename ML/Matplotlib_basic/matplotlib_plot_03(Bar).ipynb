```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
%matplotlib inline
```

```python
# BAr chart or BAr graph : revenue present
company = ["google","amazon","msft","fb"]
revenue = [90,136,89,27]
plt.xlabel("company")
plt.ylabel("revenue")
plt.title("us company revenue")
plt.bar(company,revenue, label = "revenue")
plt.legend()
```

```python
# represent two bar in one 
profit = [40,5,38,12]
plt.xlabel("company")
plt.ylabel("revenue")
plt.title("us company revenue")
plt.bar(company,revenue, label = "revenue")
plt.bar(company,profit, color="red", label="profit")
plt.legend()
```

```python
# change position 2nd bar chart
# at first change string into float or integer using np.arange(le)
xpos = np.arange(len(company))
plt.xlabel("company")
plt.ylabel("revenue")
plt.title("us company revenue")
plt.xticks(xpos,company)
plt.bar(xpos-0.2,revenue, label = "revenue")
plt.bar(xpos+0.2,profit, color="red", label="profit")
plt.legend()
```

```python
# want to show bar chart horizontally?
ypos = np.arange(len(company))
plt.xlabel("revenue")
plt.ylabel("company")
plt.title("us company revenue")
plt.yticks(ypos,company)
plt.barh(ypos-0.2,revenue, label = "revenue")
plt.barh(ypos+0.2,profit, color="red", label="profit")
plt.legend()
```
