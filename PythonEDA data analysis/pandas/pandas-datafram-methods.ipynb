```python
# using lists
student_data = [
    [100,80,10],
    [90,70,7],
    [120,100,14],
    [80,50,2]
]
# fisrt columns name= iq[100,90,120,80], 2nd columns name=marks[80,70,100,50], 
# third columns name = package[10,7,14,2]
stdData = pd.DataFrame(student_data,columns=['iq','marks','package'])
stdData
```

```python
# for single data i can use series not dataframe. 
std = [100,80,10]
pd.Series(std,name='students marks')
```

```python
# using dicts

student_dict = {
    'name':['nitish','ankit','rupesh','rishabh','amit','ankita'],
    'iq':[100,90,120,80,0,0],
    'marks':[80,70,100,50,0,0],
    'package':[10,7,14,2,0,0]
}

students = pd.DataFrame(student_dict)
# set index from coulmns: name or iq or marks or package
students.set_index('name',inplace=True)
students
```
